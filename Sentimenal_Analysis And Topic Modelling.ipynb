```python
dir(senti)
```

```python
def sentimentalAnalysis(data,columnname):
    data['scores'] = data[columnname].apply(lambda text: senti.polarity_scores(text))
    data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
    data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
    data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
    data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
    "Creating column for positive ,Negative using compound score"
    data['Positive/Negative'] = data['compound'].apply(lambda v: 'pos' if v >=0 else 'neg')
    Positive_Negative=data['Positive/Negative']
    return Positive_Negative,data
```

```python
Positive_Negative,data_Sentiment=sentimentalAnalysis(dataset,columnname='Chat')
```

# Topic Modelling

```python
# 1)IN sentimental analysis , we didn't done preprocessin because in sentimental analysis class itself have preprocessing techniques
# 2) But In Topic Modelling, We have to do pre processing(counter vector, tfid vector, hash vector)
# 3)For Topic Modelling-> two algorithms are there 1. NMF and 2.LDA
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#max_df=0.95-> remove most common words,min_df=2->remove most rare words
tfid_preprocess=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
tfid_fit=tfid_preprocess.fit_transform(data_Sentiment["Chat"])
```

```python
tfid_fit
```

```python
from sklearn.decomposition import NMF
Nmf_model=NMF(n_components=5,random_state=42)#n_components=5-> 5 Topics
Nmf_model.fit(tfid_fit)
```

```python
print(Nmf_model.components_)#-> matrix of shape [n_topics, n_features]  Each row has a topic and contains the importance (weight) of each word (feature) in that topic.
#each row is a topics and column is a word
```

```python
feature_names = tfid_preprocess.get_feature_names_out()
for index,topics in enumerate(Nmf_model.components_):
    topic_keywords = {}
    Top_Words=([feature_names[i] for i in topics.argsort()[-10:]])
    print(f"Topic #{index}: {Top_Words}")
    #topic_keywords[index] = ", ".join(Top_Words)
    #data_Sentiment["Topic_Keywords"] = data_Sentiment["Topic"].map(topic_keywords)
```

```python
topic_results=Nmf_model.transform(tfid_fit)
data_Sentiment["Topic"]=topic_results.argmax(axis=1)
```
